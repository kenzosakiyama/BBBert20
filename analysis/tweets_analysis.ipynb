```python
import pandas as pd
```

```python
PATH="data/paredao5/bianca.csv"
```

```python
cand_df = pd.read_csv(PATH)
```

# Positivos cr√≠ticos

An√°lise de tweets classificados como positivos por√©m com score de classifica√ß√£o (softmax) baixo.

```python
aux = cand_df["sentiment"] == "Positivo"
pos_df = cand_df[aux]
pos_df = pos_df[["tweet", "positive_score", "mentions", "hashtags"]]
```

```python
pd.options.display.max_colwidth = 1000
pos_df.sort_values(by="positive_score").iloc[:50]
```
