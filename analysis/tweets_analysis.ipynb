# An√°lise de hashtags

```python
raw_hashtags = cand_df["hashtags"].values
```

```python
def get_hashtags(df: pd.DataFrame):
    "Fun√ß√£o para extrair hashtags de uma coluna com strings no formato: '['hashtag1', 'hashtag2']' "

    all_hashtags = []
    for i in range(len(df)):
        raw = df["hashtags"].iloc[i][1:-1]
        if len(raw)  == 0: continue
        hashtags = [tag_to_clean.strip()[1:-1] for tag_to_clean in raw.split(",")]
        
        all_hashtags.extend(hashtags)
    
    return all_hashtags
```

```python
hashtags_df = pd.DataFrame(get_hashtags(cand_df), columns=["hashtag"])
```
