```python
import pandas as pd
import numpy as np
```

```python
CAND1_PATH = "prior.csv"
CAND2_PATH = "hadson.csv"
```

```python
candidate1_df = pd.read_csv(CAND1_PATH)
candidate2_df = pd.read_csv(CAND2_PATH)
```

```python
cand1 = candidate1_df[["tweet", "sentiment"]]
cand2 = candidate2_df[["tweet", "sentiment"]]
```

# Estatísticas

```python
candidates = {"prior": cand1, "hadson": cand2}
```

```python
statistics = {"nome": [], "positivos": [], "neutros": [], "negativos": []}
for name in candidates.keys():
    current_df = candidates[name]
    n_tweets = len(current_df)
    statistics["nome"].append(name)
    statistics["positivos"].append((current_df["sentiment"] == "Positivo").sum()/n_tweets)
    statistics["neutros"].append((current_df["sentiment"] == "Neutro").sum()/n_tweets)
    statistics["negativos"].append((current_df["sentiment"] == "Negativo").sum()/n_tweets)
```

```python
stats_df = pd.DataFrame.from_dict(statistics).set_index("nome")
```
